```python
import pandas as pd
```

```python
stud_details = {'stu_id':[1,2,3,4,5],
               'Stu_name':['A','B','C','D','E'],
               'Region':['N','S','S','W','E']}
```

```python
stud_details
```

```python
student_details = pd.DataFrame(stud_details)
```

```python
type(student_details)
```

```python
stud_marks = {'st_id':[1,2,3,4,6],
             'subject':['Py','Tab','R','Py','R'],
             'Marks':[34,43,23,33,54]}
```

```python
student_marks = pd.DataFrame(stud_marks)
```

```python
student_details
```

```python
student_marks
```

```python
pd.merge(student_details, student_marks, left_on='stu_id', right_on='st_id',how='inner')
```

```python
pd.merge(student_details, student_marks, left_on='stu_id', right_on='st_id',how='left')
```

```python
pd.merge(student_details, student_marks, left_on='stu_id', right_on='st_id',how='right')
```

```python
pd.merge(student_details, student_marks, left_on='stu_id', right_on='st_id',how='outer')
```
